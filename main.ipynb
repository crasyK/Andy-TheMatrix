```python
from PIL import Image, ImageDraw, ImageFont, ImageFilter
import matplotlib.pyplot as plt
import random
import string

# Generate the same type of random string grid
string_grid = ['\n'.join([' '.join(random.choices(string.ascii_letters + string.digits + string.punctuation, k=53)) for _ in range(28)]) for _ in range(6)]
grid_dimensions = {'h': len(string_grid[0].split('\n')), 'w': len(max(string_grid[0].split('\n'), key=len))+1}
print(grid_dimensions)
print(string_grid[0])
```

```python
html_colors = ["#FF0000", "#ff8700", "#ffd300", "#deff0a", "#a1ff0a", "#39ff14", "#00f59b", "#0aff99", "#0aefff", "#147df5", "#580aff", "#be0aff"]

hackclub = r"""$$\   $$\  $$$$$$\   $$$$$$\  $$\   $$\  $$$$$$\  $$\      $$\   $$\ $$$$$$$\  
$$ |  $$ |$$  __$$\ $$  __$$\ $$ | $$  |$$  __$$\ $$ |     $$ |  $$ |$$  __$$\ 
$$ |  $$ |$$ /  $$ |$$ /  \__|$$ |$$  / $$ /  \__|$$ |     $$ |  $$ |$$ |  $$ |
$$$$$$$$ |$$$$$$$$ |$$ |      $$$$$  /  $$ |      $$ |     $$ |  $$ |$$$$$$$\ |
$$  __$$ |$$  __$$ |$$ |      $$  $$<   $$ |      $$ |     $$ |  $$ |$$  __$$\ 
$$ |  $$ |$$ |  $$ |$$ |  $$\ $$ |\$$\  $$ |  $$\ $$ |     $$ |  $$ |$$ |  $$ |
$$ |  $$ |$$ |  $$ |\$$$$$$  |$$ | \$$\ \$$$$$$  |$$$$$$$$\\$$$$$$  |$$$$$$$  |
\__|  \__|\__|  \__| \______/ \__|  \__| \______/ \________|\______/ \_______/ """

art_dimensions = {'h': len(hackclub.split('\n')), 'w': len(max(hackclub.split('\n'), key=len))}

print(f"ASCII Art Dimensions: w:{art_dimensions['w']}xh:{art_dimensions['h']}")

for j in range(len(string_grid)):
    for i in range(art_dimensions['h']):
        U = (grid_dimensions['w'] - art_dimensions['w']) // 2 +(grid_dimensions['w'])*((grid_dimensions['h'] - art_dimensions['h']) // 2)
        current_line = string_grid[j][U+ i*(grid_dimensions['w']): U+ (i)*(grid_dimensions['w']) + art_dimensions['w']]
        string_grid[j] = string_grid[j].replace(current_line, art_dimensions['w']*' ')#hackclub.split('\n')[i]
        

print(string_grid[0])
```

```python
# Signature
signature = "@nedimark, 2025"
string_grid[-1] = string_grid[-1].replace(string_grid[-1][-len(signature):], signature)
```

```python
img = Image.new('RGB', (1800, 800), 'black')
draw = ImageDraw.Draw(img)

font = ImageFont.truetype("/usr/share/fonts/truetype/ubuntu/UbuntuMono-BI.ttf")

colors = random.sample(html_colors,12)  # Green color for the text

print(colors)
color_count = 0
for i in range(2):
    # Draw each string grid on the image
    for j in range(3):
        draw.text((10 + j * 600, 10 + i * 400), string_grid[color_count], fill=colors[color_count], font=font)
        draw.text((85 + j * 600, 148 + i * 400 ), hackclub, fill=colors[-color_count+1], font=font)
        color_count += 1

plt.figure(figsize=(24, 16))
plt.imshow(img)
plt.axis('off')
plt.show()
```
